```python
import numpy as np
from matplotlib import pyplot as plt

def gradientDescent(a, b, c, leftX, rightX, alpha, w):
    x = np.arange(leftX, rightX, 0.1)
    y = (a/2)*x**2 + b*x + c
    wStar = -b/a
    wt = w
    difference = []

    plt.plot(x, y, color = "red")
    plt.scatter(-b/a, c - b**(2)/(2*a), color = "blue", marker = '*')
    
    for i in range(10):
        plt.scatter(wt, (a/2)*(wt)**2 + b*(wt) + c,color = "black", marker = 's')
        wt = w - alpha*(a*w + b) 
        difference.append(wt - w)
        w = wt
    
    plt.show()
    plt.plot(difference)
    plt.show()
```

```python
a = 1 
b = 0 
c = 0
leftX = -10
rightX = 10
alpha = 0.5
w0 = +10.5
gradientDescent(a, b, c, leftX, rightX, alpha, w0 )
alpha = 1.69
gradientDescent(a, b, c, leftX, rightX, alpha, w0 )
alpha = -0.2
gradientDescent(a, b, c, leftX, rightX, alpha, w0 )
a = 1
b = 1
c = 1
alpha = -1
gradientDescent(a, b, c, leftX, rightX, alpha, w0 )
```
