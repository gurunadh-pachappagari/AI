```python
import numpy as np
from matplotlib import pyplot as plt

def drawQuadratic(a, b, c, leftX, rightX):
    x = np.arange(leftX, rightX, 0.1)
    y = (a/2)*x**2 + b*x + c
#     print(x, y)
    plt.plot(x, y, color = "red")
    plt.scatter(-b/a, c - b**(2)/(2*a), color = "blue", marker = '*')
    plt.show()
    
    
a = 1 
b = 0 
c = 0
leftX = -10
rightX = 10
drawQuadratic(a, b, c, leftX, rightX)
a = 0.1 
b = 0 
c = 0
leftX = -10
rightX = 10
drawQuadratic(a, b, c, leftX, rightX)
a = 10 
b = 0 
c = 0
leftX = -10
rightX = 10
drawQuadratic(a, b, c, leftX, rightX)
a = 1 
b = 0 
c = 10
leftX = -10
rightX = 10
drawQuadratic(a, b, c, leftX, rightX)
a = 1 
b = 1 
c = 1
leftX = -10
rightX = 10
drawQuadratic(a, b, c, leftX, rightX)
a = 0.1 
b = -1 
c = -1
leftX = 0
rightX = 20
drawQuadratic(a, b, c, leftX, rightX)

```
