```python
import numpy as np
from matplotlib import pyplot as plt
x = np.linspace(-15, 10, 30)
y = np.linspace(-10, 10, 30)
x, y = np.meshgrid(x, y)
# print(type(x))
z1 = x**2 + y**2 + 5*x - 3*y - 2
z2 = 10*x**2 + y**2
dx,dy = np.gradient(z1)
print(z1)
plt.quiver(x, y, dx, dy)
plt.contour(x, y, z1)
plt.show()


dx,dy = np.gradient(z2)
plt.quiver(x, y, dx, dy)
plt.contour(x, y, z2)
plt.show()
```
